```python
import cv2
import matplotlib.pyplot as plt
img=cv2.imread('fr.jpeg')
```

```python
plt.imshow(img)
plt.show()
```

```python
detect = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray)
cv2.imwrite('jr.jpeg',img) 
```

```python
face = detect.detectMultiScale(gray, 1.25, 3)
for (x,y,w,h) in face:
    cv2.rectangle(img,(x,y), (x+w, y+h), (0,255),2)
plt.imshow(img)
```
